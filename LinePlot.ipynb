### Multi-line plot

```python
# libraries and data
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# Make a data frame
df=pd.DataFrame({'x': range(1,11), 'y1': np.random.randn(10), 'y2': np.random.randn(10)+range(1,11), 'y3': np.random.randn(10)+range(11,21), 'y4': np.random.randn(10)+range(6,16)})
 
# style
plt.style.use('seaborn-darkgrid')
 
# create a color palette

 
# multiple line plot
num=0
for column in df.drop('x', axis=1):
    num+=1
    plt.plot(df['x'], df[column], marker='', color=sns.color_palette("Set2")[num], linewidth=2, alpha=0.9, label=column)
 
# Add legend
plt.legend(loc=2, ncol=2)
 
# Add titles
plt.title("A Spaghetti plot", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Time")
plt.ylabel("Score")
```

### Multi-line with highlight

```python
# libraries and data
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# Make a data frame
df=pd.DataFrame({'x': range(1,11), 'y1': np.random.randn(10), 'y2': np.random.randn(10)+range(1,11), 'y3': np.random.randn(10)+range(11,21), 'y4': np.random.randn(10)+range(6,16)})
 
#plt.style.use('fivethirtyeight')
plt.style.use('seaborn-darkgrid')
my_dpi=96
plt.figure(figsize=(480/my_dpi, 480/my_dpi), dpi=my_dpi)
 
# multiple line plot
for column in df.drop('x', axis=1):

    plt.plot(df['x'], df[column], marker='', color=sns.color_palette("Blues")[2], linewidth=1.5, alpha=0.6)
   
 
# Now re do the interesting curve, but biger with distinct color
plt.plot(df['x'], df['y4'], marker='', color='red', linewidth=4, alpha=0.7)

plt.scatter(x=df['x'], y=df['y4'], s=40, color='red')
# Change xlim
plt.xlim(0,12)
 
# Let's annotate the plot
num=0
for i in df.values[9][1:]:
   num+=1
   name=list(df)[num]
   if name != 'y4':
      plt.text(10.2, i, name, horizontalalignment='left', size='small', color=sns.color_palette("Blues")[2])
 
# And add a special annotation for the group we are interested in
plt.text(10.2, df.y4.tail(1), 'Mr Big Shot', horizontalalignment='left', size='small', color='red')
 
# Add titles
plt.title("Evolution of Mr Big shot vs other students", loc='left', fontsize=12, fontweight=0, color='k')
plt.xlabel("Time")
plt.ylabel("Score")
```

### Multi line plot

```python
# Initialize the figure
plt.style.use('seaborn-darkgrid')
 
# create a color palette
palette = plt.get_cmap('Set1')
 
# multiple line plot
num=0
for column in df.drop('x', axis=1):
    num+=1
 
    # Find the right spot on the plot
    plt.subplot(2,2, num)
 
    # Plot the lineplot
    plt.plot(df['x'], df[column], marker='', color=palette(num), linewidth=1.9, alpha=0.9, label=column)
 
    # Same limits for everybody!
    plt.xlim(0,10)
    plt.ylim(-2,22)
 
    # Not ticks everywhere
    if num in range(3) :
        plt.tick_params(labelbottom=False)
    if num not in [1, 3] :
        plt.tick_params(labelleft=False)
 
    # Add title
    plt.title(column, loc='left', fontsize=12, fontweight=0, color=palette(num) )
 
# general title
plt.suptitle("How the 4 students improved\nthese past few days?", fontsize=13, fontweight=0, color='black', style='italic', y=1.02)
 
```

### Multiple line plot with highlights

```python
# Initialize the figure
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(8,8))
# create a color palette
palette = plt.get_cmap('Set1')
 
# multiple line plot
num=0
for column in df.drop('x', axis=1):
    num+=1
 
    # Find the right spot on the plot
    plt.subplot(2,2, num)
 
    # plot every groups, but discreet
    for v in df.drop('x', axis=1):
        plt.plot(df['x'], df[v], marker='', color=sns.color_palette("Blues")[3], linewidth=1, alpha=0.5)
 
    # Plot the lineplot
    plt.plot(df['x'], df[column], marker='', color=palette(num), linewidth=2.4, alpha=0.9, label=column)
 
    # Same limits for everybody!
    plt.xlim(0,10)
    plt.ylim(-2,22)
 
    # Not ticks everywhere
    if num in range(3) :
        plt.tick_params(labelbottom=False)
    if num not in [1,3] :
        plt.tick_params(labelleft=False)
 
    # Add title
    plt.title(column, loc='left', fontsize=12, fontweight=0, color=palette(num) )
 
# general title
plt.suptitle("How the 4 students improved\nthese past few days?", fontsize=13, fontweight=0, color='black', style='italic', y=1.02)
```

## Parallel Line plot

```python
# libraries
import pandas
import matplotlib.pyplot as plt
from pandas.tools.plotting import parallel_coordinates
 
# Take the iris dataset
import seaborn as sns
data = sns.load_dataset('iris')
 
# Make the plot
parallel_coordinates(data, 'species', colormap=plt.get_cmap("Set2"))
plt.show()
```
